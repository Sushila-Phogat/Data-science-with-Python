```
#find out common merchants
merclist=[]
for i in data[(data['fraud'].astype(str).str.strip()=='True') & (data['date'].astype(str).str.strip()<min(data['date'].astype(str).str.strip()))][['account','date']]:
  merclist.append(data[(data['account'].astype(str).str.strip()==i)]['merchant'].unique())

comnlist=merclist[0]
for i in range(len(merclist)):
   comnlist1=(list(set(merclist[i]).intersection(comnlist)))
   comnlist=comnlist1

comnlist
```
